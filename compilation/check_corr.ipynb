```python
import pandas as pd
```

```python
similarities_path = "/home/milte/Desktop/AI-nspired/compilation/csv"
```

```python
mean_sims_web = dict()
mean_sims_ai = dict()

for i in range(1,28):
    sims = pd.read_csv(f"{similarities_path}/{i}.csv")
    mean_sims_web[i] = sims[sims["image2"].str.contains("web")]["resnet cosine similarity"].mean()
    mean_sims_ai[i] = sims[sims["image2"].str.contains("ai")]["resnet cosine similarity"].mean()

pd.DataFrame({"web mean similarity":[mean_sims_web[i] for i in range(1,28)], "ai mean similarity":[mean_sims_ai[i] for i in range(1,28)]},index=list(range(1,28)))
```

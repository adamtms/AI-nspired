```python
import pandas as pd
```

```python
similarities_path = "/home/milte/Desktop/AI-nspired/compilation/csv"
```

```python
mean_sims_web = dict()
mean_sims_ai = dict()

for i in range(1,28):
    sims = pd.read_csv(f"{similarities_path}/{i}.csv")
    
    sims_web = sims[sims["image2"].str.contains("web")]["resnet cosine similarity"].sort_values(ascending=False)
    sims_ai = sims[sims["image2"].str.contains("ai")]["resnet cosine similarity"].sort_values(ascending=False)
    
    weights_web = [1/((1+x)**0.5) for x in range(len(sims_web))]
    weights_ai = [1/((1+x)**0.5) for x in range(len(sims_ai))]
    
    weighted_mean_web = sum(sims_web * weights_web) / sum(weights_web) if sum(weights_web) != 0 else None
    weighted_mean_ai = sum(sims_ai * weights_ai) / sum(weights_ai) if sum(weights_ai) != 0 else None
    
    mean_sims_web[i] = weighted_mean_web
    mean_sims_ai[i] = weighted_mean_ai

pd.DataFrame({"web mean similarity":[mean_sims_web[i] for i in range(1,28)], "ai mean similarity":[mean_sims_ai[i] for i in range(1,28)]},index=list(range(1,28)))
```

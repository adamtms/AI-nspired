```python
import pandas as pd
```

```python
best_image = pd.read_csv("data/Initial_Survey.csv").iloc[:, -2]
list_ = []
for i in best_image.to_list():
    try:
        list_.append("AI" if int(i) > 10 else "Web")
    except ValueError:
        print(i)
pd.Series(list_).value_counts()
# + 2 AI, + 4 Web
# total 31 AI, 19 Web
```

```python
summary = pd.read_csv("data/Summary_Survey.csv")
summary = summary.iloc[:, 3:-1].map(lambda x: "AI" if x >= 11 else "Web").apply(pd.Series.value_counts).fillna(0).astype(int).T
summary.index = ["EFEKT ZASKOCZENIA", "INNOWACYJNOŚĆ ROZWIĄZAŃ", "JAKOŚĆ WIZUALNA", "NASTRÓJ SCENY", "KOLORYSTYKA", "ZESTAWIENIE MATERIAŁOWE", "CAŁOKSZTAŁT"]
summary
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
summary.plot.barh(ax=ax, title="Preferencje wizualne AI vs Web", legend=False)
fig.legend(["AI", "Web"], loc="right")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

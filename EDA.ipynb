```python
def fix_name(name: pd.DataFrame):
    name = name.strip().lower().split()
    name.sort()
    return ' '.join(name)
```

```python
# People with typo:
#   - daria kachan
#   - oliwia wróblrewska
#   - martyna szpryt
#   - ema sauczuk
#   - daria kachan
#   - daria tiasko
# People who doesn't send surveys:
#   - kije weronika
#   - mądrawska oliwia
# People who doesn't send only Initial Survey:
#   - eliza ratajczak
# Marta Kowalska send two Initial Surveys
# Michalak Urszula send Initial Survey, but doesn't apear in participants list

participants["Fixed Name"] = participants["Name"].apply(fix_name)
initial_survey["Fixed Name"] = initial_survey["1. Imię i Nazwisko"].apply(fix_name)
summary_survey["Fixed Name"] = summary_survey["1. Imię i Nazwisko"].apply(fix_name)
print(f"Participants shape: {participants.shape}")
print(f"Initial Survey shape: {initial_survey.shape}")
print(f"Summary Survey shape: {summary_survey.shape}")
all = pd.merge(participants, initial_survey, on="Fixed Name", how="inner")
print("Size after merge Participants and Initial Survey: ", all.shape)
all = pd.merge(all, summary_survey, on="Fixed Name", how="inner")
print("Size after merge Participants, Initial Survey and Summary Survey: ", all.shape)
print("Not matched names from Participants:")
print(participants.loc[~participants["Fixed Name"].isin(all["Fixed Name"]), ["Fixed Name"]])
print()
print("Not matched names from Initial Survey:")
print(initial_survey.loc[~initial_survey["Fixed Name"].isin(all["Fixed Name"]), ["Fixed Name"]])
print()
print("Not matched names from Summary Survey:")
print(summary_survey.loc[~summary_survey["Fixed Name"].isin(all["Fixed Name"]), ["Fixed Name"]])
print()
```
